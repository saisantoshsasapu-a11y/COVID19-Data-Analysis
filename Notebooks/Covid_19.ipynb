```python
# Example: sum confirmed and recovered cases by country
country_summary = data.groupby('Country/Region')[['Confirmed', 'Recovered', 'Deaths']].sum()
print(country_summary.head())
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
top_countries = country_summary.sort_values('Confirmed', ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries['Confirmed'])
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.show()
```
